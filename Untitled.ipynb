```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv('/home/madan/Artificial Intelligence/DeepLeaarning/Multiclass Classification Of Flower Species/iris.data')
df.head()
```

```python
dataset=df.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
```

```python
print(X,Y)
```

```python
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
dummy_y
```

```python
df.values[:,:4]
```

```python
for i in range(len(df.values[:,4])):
    if df.values[:,4][i]=='Iris-versicolor':
        df.values[:,4][i]=1
    elif df.values[:,4][i]=='Iris-virginica':
        df.values[:,4][i]=2
    else:
        df.values[:,4][i]=0
plt.plot(df.values[:,0],df.values[:,4],marker='o')
plt.plot(df.values[:,1],df.values[:,4],marker='p')
plt.plot(df.values[:,2],df.values[:,4],marker='*')
plt.plot(df.values[:,3],df.values[:,4],marker='o')
plt.rcParams["font.size"] = "50"
plt.show()
```

```python
# define Use model
def Use_model():
    # create model
    model = Sequential()
    model.add(Dense(40, input_dim=4, init= 'normal' , activation= 'relu' ))
    model.add(Dense(12,init='normal',activation='relu'))
    model.add(Dense(3, init= 'normal' , activation= 'sigmoid' ))
    # Compile model
    model.compile(loss= 'categorical_crossentropy' , optimizer= 'adam' , metrics=[ 'accuracy' ])
    return model
```

```python
#make a reference for keras classifier
estimator = KerasClassifier(build_fn=Use_model, nb_epoch=1000, batch_size=5)
```

```python
seed=10
np.random.seed(seed)
#use kfold validation for data validating
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
#calculate the result with accuracy and print in percentage form
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
